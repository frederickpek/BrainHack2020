```python
import json

json_file = open("D:/CS/Python/BrainHack2020/CV/til2020trainingdata/val.json", "r")
val_json = json.load(json_file)
json_file.close()

json_file = open("D:/CS/Python/BrainHack2020/CV/til2020trainingdata/CV_interim_evaluation.json", "r")
interim_json = json.load(json_file)
json_file.close()

print(val_json.keys())
print(interim_json.keys())

print("Annotations: ", len(val_json["annotations"]))

IDs = []
# dict_keys(['area', 'iscrowd', 'id', 'image_id', 'category_id', 'bbox'])
for d in val_json["annotations"]:
    IDs.append(d["image_id"])

IDs = list(set(IDs))
IDs.sort()
len(IDs[::-1])

    
# 'area' = 0,
# 'iscrowd' = 0, 
# 'id'->1 to len(images), //dictates the number of individial annotations 
# 'image_id'-> 1 to len(images) number of IDs, //can have multiple of same ones, must match image_name.jpg
# 'category_id' -> 0, 
# 'bbox' -> [0, 0, 0, 0]

val_json["images"]
val_json["categories"]
```

```python
import os
test_tolder = os.path.join( "D:/CS/Python/BrainHack2020/CV/til2020trainingdata/val/val" )
test_files = list(os.listdir( test_tolder ))
print(len(test_files))
# test_files[x][:-4] to strip the ".jpg"
```

```python
annotations = []
for x in range(len(test_files)):
    image_id = test_files[x]
    annotation = {'area':0, 
                  'iscrowd':0, 
                  'id':x+1, 
                  'image_id':image_id[:-4], 
                  'category_id':1, 
                  'bbox':[0, 0, 0, 0]}
    annotations.append(annotation)
print(len(annotations))
```

```python
dummy_val_dict = {}
dummy_val_dict["annotations"] = annotations
dummy_val_dict["images"] =val_json["images"]
dummy_val_dict["categories"] = val_json["categories"]
dummy_val_dict
```

```python
dummy_annotation_val = open("D:/CS/Python/BrainHack2020/CV/til2020trainingdata/dummy_val.json", "w")
json.dump(dummy_val_dict, dummy_annotation_val)
```
